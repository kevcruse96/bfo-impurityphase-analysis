```python
import pandas as pd
import numpy as np
import json

from pprint import pprint

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('./data/BFO_film_synth_df_codified_20211101.csv')
#populate missing MOE_EG ratios
MOE_EG_mean = df[df['MOE_EG_ratio']>1]['MOE_EG_ratio'].mean()
for i, (e, m, ratio) in enumerate(
    zip(df['ethylene_glycol'].to_list(), 
        df['2-methoxyethanol'].to_list(), 
        df['MOE_EG_ratio'].to_list())
):
    if e and m and not ratio:
        df['MOE_EG_ratio'].iat[i] =  MOE_EG_mean
        
#populate missing MOE_acetic ratios
MOE_acetic_mean = df[df['MOE_acetic_ratio']!=0]['MOE_acetic_ratio'].mean()
for i, (m, a, ratio) in enumerate(
    zip(df['2-methoxyethanol'].to_list(), 
        df['acetic_acids'].to_list(),
        df['MOE_acetic_ratio'].to_list())
):
    if a and m and not ratio:
        df['MOE_acetic_ratio'].iat[i] =  MOE_acetic_mean
        
#populate missing EG_acetic ratios
for i, (e, a) in enumerate(
    zip(df['ethylene_glycol'].to_list(), 
        df['acetic_acids'].to_list())
):
    if a and m:
        df['EG_acetic_ratio'].iat[i] =  1

#populate missing mixing time, mixing temp, annealing time, pH, prebake temp, and precursor concentration values
df['bi_fe_ratio'] = df['bi_fe_ratio'].fillna(df['bi_fe_ratio'].median())
df['stirring_time_hr'] = df['stirring_time_hr'].fillna(df['stirring_time_hr'].median())
df['stirring_temp_degC'] = df['stirring_temp_degC'].fillna(df['stirring_temp_degC'].median())
df['annealing_time_hr'] = df['annealing_time_hr'].fillna(df['annealing_time_hr'].median())
df['pH'] = df['pH'].fillna(df['pH'].mean())
df['prebake_degC'] = df['prebake_degC'].fillna(0)
for i, (p, t) in enumerate(zip(df['prebake_degC'].to_list(), df['prebake_time_hr'].to_list())):
    if p and not t:
        t = df['prebake_time_hr'].median()
df['prebake_time_hr'] = df['prebake_time_hr'].fillna(0)
df['age_days'] = df['age_days'].fillna(0)
df['age_temp_degC'] = df['age_temp_degC'].fillna(0)
df['precursor_concentration'] = df['precursor_concentration'].fillna(df['precursor_concentration'].median())

refined_df = df.drop([
    'doi',
    'ethylene_glycol',
    '2-methoxyethanol',
    'acetic_acids',
    'citric_acid',
    'pH',
    'separate_hydrolysis'
], axis=1)

refined_df.to_csv('./data/BFO_film_synth_df_refined_20211101.csv', index=False)

display(refined_df)
```
